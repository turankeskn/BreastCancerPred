## Outlier Handling

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3-Q1


min_bound = Q1 - 1.5 * IQR
max_bound = Q3 + 1.5 * IQR


df_out = df[~((df < (min_bound)) | (df > (max_bound))).any(axis=1)]
print("Before:",df.shape, "After:",df_out.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.preprocessing import MinMaxScaler
```

```python
scale = MinMaxScaler()
le = LabelEncoder()

X = df_out.drop("diagnosis" ,axis = 1)
y = le.fit_transform(df_out["diagnosis"])

X = scale.fit_transform(X)
X
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y ,test_size = 0.25, random_state = 42)

print("X Test: ",X_test.shape)
print("X Train:",X_train.shape)
print("Y Test: ",y_test.shape)
print("Y Train: ",y_train.shape)
```

## Logistic Regression

```python
#Model
lr = LogisticRegression()
lr_model = lr.fit(X_train,y_train)

#Prediction
lr_pred = lr_model.predict(X_test)
train_pred = lr_model.predict(X_train)
```

```python
print("X Test: ",classification_report(y_test,lr_pred))

print("X Train: ",classification_report(y_train,train_pred))

plt.figure(figsize=(8,8))
sns.heatmap(confusion_matrix(y_test,lr_pred),annot=True)
```

## KNN

```python
#Model
knn = KNeighborsClassifier()
knn_model = knn.fit(X_train,y_train)

#Prediction
knn_pred = knn_model.predict(X_test)
knn_train_pred = knn_model.predict(X_train)
```

```python
print("X Test: ",classification_report(y_test,knn_pred))

print("X Train: ",classification_report(y_train,knn_train_pred))

plt.figure(figsize=(8,8))
sns.heatmap(confusion_matrix(y_test,knn_pred),annot=True)
```

## XGBClassifier

```python
xgb = XGBClassifier(learning_rate = 0.1,
                   max_depth = 6,
                   min_samples_split = 2,
                   n_estimators = 50,
                   subsample = 0.8)
xgb_model = xgb.fit(X_train,y_train)

#Prediction
xgb_pred = xgb_model.predict(X_test)
xgb_train_pred = xgb_model.predict(X_train)
```

```python
print("X Test: ",classification_report(y_test,xgb_pred))

print("X Train: ",classification_report(y_train,xgb_train_pred))

plt.figure(figsize=(8,8))
sns.heatmap(confusion_matrix(y_test,xgb_pred),annot=True)
```

## Random Forest

```python
rf = RandomForestClassifier()
rf_params = {'max_features': ('auto','sqrt'),
             'n_estimators': [50, 10, 100],
             'max_depth': [5, 10, None],
             'min_samples_split': [5, 10, 15],
             'min_samples_leaf': [1, 2, 5, 10],
             'bootstrap': [True, False]}

rf_model_cv = GridSearchCV(rf,rf_params,cv=5,n_jobs=-1)
rf_model_cv.fit(X_train,y_train)
rf_model_cv.best_params_
```

```python
rf_tuned = RandomForestClassifier(max_depth = None,
                                 max_features = "auto",
                                 n_estimators = 10,
                                 bootstrap = True,
                                 min_samples_leaf = 1,
                                 min_samples_split = 10).fit(X_train,y_train)

#Prediction
rf_pred = rf_tuned.predict(X_test)
rf_train_pred = rf_tuned.predict(X_train)
```

```python
print("X Test: ",classification_report(y_test,rf_pred))

print("X Train: ",classification_report(y_train,rf_train_pred))

plt.figure(figsize=(8,8))
sns.heatmap(confusion_matrix(y_test,rf_pred),annot=True)
```

## Random Forest Features Selection

```python
columns = df.iloc[:,1:].columns
data = pd.DataFrame(X,columns = columns)
```

```python
feature_imp = pd.Series(rf_tuned.feature_importances_, index= data.columns)
feature_imp.nlargest(10).plot(kind='barh');
```

```python
best_feat = feature_imp.nlargest(10).index.to_list()

X_reduced = data[feature_imp.nlargest(10).index]

Xr_train, Xr_test, yr_train, yr_test = train_test_split(X_reduced, y, test_size = 0.25, random_state = 42)
```

```python
rf_params = {'max_features': ('auto','sqrt'),
             'n_estimators': [50, 10, 100],
             'max_depth': [5, 10, None],
             'min_samples_split': [5, 10, 15],
             'min_samples_leaf': [1, 2, 5, 10],
             'bootstrap': [True, False]}

rf_reduced = GridSearchCV(rf,rf_params,cv=5,n_jobs=-1)
rf_reduced.fit(Xr_train,yr_train)
rf_reduced.best_params_
```

```python
rff_reduced = RandomForestClassifier(max_features = "auto",
                                      n_estimators = 100,
                                      bootstrap = True,
                                      max_depth = None,
                                      min_samples_leaf = 1,
                                      min_samples_split = 5).fit(Xr_train,yr_train)


rff_pred = rff_reduced.predict(Xr_test)
rff_train_pred = rff_reduced.predict(Xr_train)
```

```python
print("X Test: ",classification_report(y_test,rf_pred))

print("X Train: ",classification_report(y_train,rf_train_pred))

plt.figure(figsize=(8,8))
sns.heatmap(confusion_matrix(y_test,rff_pred),annot=True)
```
